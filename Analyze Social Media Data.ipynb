```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
socialMedia = pd.read_csv("C:/Users/Dell/Documents/python3/social_media_usage/social_media_usage.csv")
```

```python
socialMedia.fillna(0, inplace=True)
```

```python
socialMedia.info()
```

```python
socialMedia.head()
```

```python
tiktok = socialMedia.loc[socialMedia['App'] == "TikTok"].count().iloc[0]
facebook = socialMedia.loc[socialMedia['App'] == "Facebook"].count().iloc[0]
linkedin = socialMedia.loc[socialMedia['App'] == "LinkedIn"].count().iloc[0]
pinterest = socialMedia.loc[socialMedia['App'] == "Pinterest"].count().iloc[0]
print(pinterest)

x = [tiktok, facebook, pinterest, linkedin]
```

# Pie Chart With Multi-Color Representation

```python
labels = ["Tiktok", "Facebook", "Pinterest", "LinkedIn"]
explode = [.2, .2, .2, .2]

range_of_colors = ["Yellow", "Blue", "Green", "Red", "Pink", "Orange", "Purple", "Grey"]

colors = []

for i in range(0,4):
    randnumber = np.random.randint(0,7)
    colors.append(range_of_colors[randnumber])

plt.pie(x, labels = labels, explode = explode, autopct= "%.2f %%", colors = colors)

plt.title("Representation of 3 selected social counts")
plt.show()
```

# Histogram

```python
plt.figure(figsize=(8,5), dpi=200)
plt.style.use('ggplot')
bins = [130, 132, 134, 136, 138, 140, 142, 146, 148, 150, 152, 154, 156, 158, 160]
plt.hist(x, label=labels, bins = bins, color = 'green')

plt.show()
```
